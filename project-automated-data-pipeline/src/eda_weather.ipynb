```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ── CHANGE THIS PATH WHEN REAL DATA IS READY ────────────────────────────────
DATA_PATH = 'data/raw/weather_data.csv'
# ────────────────────────────────────────────────────────────────────────────

CITY_COLORS = {
    'tallahassee': '#2196F3',
    'dallas':      '#F44336',
    'New York':    '#4CAF50',
}
CITY_ORDER = ['tallahassee', 'dallas', 'New York']

df = pd.read_csv(DATA_PATH, parse_dates=['date', 'run_timestamp'])
print(f'Rows   : {len(df):,}')
print(f'Cities : {sorted(df["city"].unique())}')
df.describe().round(2)
```

## Plot 1 — Daily High Temperature by City

```python
fig, ax = plt.subplots(figsize=(11, 4))

for city in CITY_ORDER:
    grp   = df[df['city'] == city]
    daily = grp.groupby('date')['seven_day_daily_temp_max'].mean().reset_index()
    ax.plot(daily['date'], daily['seven_day_daily_temp_max'],
            marker='o', linewidth=2, markersize=5,
            color=CITY_COLORS[city], label=city.title())

ax.set_title('Average Daily High Temperature by City (°F)', fontsize=13, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°F)')
ax.legend()
ax.grid(True, alpha=0.3)
fig.autofmt_xdate()
plt.tight_layout()
plt.savefig('figures/plot_daily_temp_max.png', dpi=150)
plt.show()
```

## Plot 2 — Precipitation Probability by City

```python
avg_precip = (
    df.groupby('city')['seven_day_precipitation_probability']
    .mean()
    .reindex(CITY_ORDER)
    .round(1)
)

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(
    [c.title() for c in CITY_ORDER],
    avg_precip.values,
    color=[CITY_COLORS[c] for c in CITY_ORDER],
    width=0.5, edgecolor='white'
)
ax.bar_label(bars, fmt='%.1f%%', padding=4)
ax.set_title('Average Precipitation Probability by City (%)', fontsize=13, fontweight='bold')
ax.set_ylabel('Precipitation Probability (%)')
ax.set_ylim(0, 100)
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('figures/plot_precip_probability.png', dpi=150)
plt.show()
```
