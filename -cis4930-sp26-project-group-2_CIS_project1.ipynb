```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

sns.set_theme(style="whitegrid")

# Copy dataset
df_bar = df_update.copy()

# Create categorical feature
df_bar["screen_time_category"] = np.where(
    df_bar["daily_screen_time_hours"] < 3, "Low",
    np.where(df_bar["daily_screen_time_hours"] < 7, "Medium", "High")
)


# Grouping with multiple aggregations
grouped_screen = df_bar.groupby("screen_time_category")["stress_level"].agg(["mean", "count"])
grouped_screen = grouped_screen.reset_index()
order = ["Low", "Medium", "High"]


# Plot
fig, ax = plt.subplots(figsize=(7, 5))

sns.barplot(
    data=grouped_screen,
    x="screen_time_category",
    y="mean",
    hue="screen_time_category",
    palette="coolwarm",
    legend=False,
    order=order,
    ax=ax
)
ax.set_title("Average Stress Level by Screen Time")
ax.set_xlabel("Screen Time")
ax.set_ylabel("Average Stress Level")
fig.tight_layout()


# Save figure
file_path = os.path.join("figures", "stress_by_screen_time.png")
fig.savefig(file_path, dpi=300)

plt.show()
```
